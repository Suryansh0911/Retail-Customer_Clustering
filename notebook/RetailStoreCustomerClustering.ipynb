```python
import datetime as dt
```

```python
df['Total Sum'] = df['Quantity'] * df['UnitPrice']
```

```python
df
```

```python
snapshot_date = df['InvoiceDate'].max() + dt.timedelta(days=1)
```

```python
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'InvoiceNo': 'count',
    'Total Sum': 'sum'
})
```

```python
rfm.rename(columns={'InvoiceDate': 'Recency',
                    'InvoiceNo': 'Frequency',
                    'TotalSum': 'Monetary'}, inplace=True)
```

```python
df
```

```python
rfm_log = np.log1p(rfm)
```

```python
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_log)
print(rfm_scaled)
```

```python
rfm_scaled_df = pd.DataFrame(rfm_scaled, columns=rfm.columns, index=rfm.index)
```

```python
rfm_scaled_df
```

```python
# Check the new distribution
plt.figure(figsize=(12,4))
for i, col in enumerate(['Recency', 'Frequency', 'Total Sum']):
    plt.subplot(1, 3, i+1)
    sns.histplot(rfm_scaled_df[col], kde=True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
rfm_pca = pca.fit_transform(rfm_scaled_df)
pca_df = pd.DataFrame(data = rfm_pca, columns = ['PC1', 'PC2'], index = rfm_scaled_df.index)
```

```python
pca_df
```

```python
from sklearn.cluster import KMeans
inertia = []
range_k = range(1, 11)
for k  in range_k:
  kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
  kmeans.fit(pca_df)
  inertia.append(kmeans.inertia_)

plt.plot(range_k, inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('The Elbow Method')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
pca_df['Cluster'] = kmeans.fit_predict(pca_df)

plt.figure(figsize=(10, 8))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=pca_df, palette='viridis', s=60)
plt.title('Customer Segments in PCA Space')
plt.show()
```

```python
rfm['Cluster'] = pca_df['Cluster']
cluster_profile = rfm.groupby('Cluster').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Total Sum': 'mean'
}).round(2)

print(cluster_profile)
```

```python
# Melt the normalized data for visualization
rfm_scaled_df['Cluster'] = pca_df['Cluster']
rfm_melt = pd.melt(rfm_scaled_df.reset_index(),
                    id_vars=['CustomerID', 'Cluster'],
                    value_vars=['Recency', 'Frequency', 'Total Sum'],
                    var_name='Metric',
                    value_name='Value')

plt.figure(figsize=(10, 6))
sns.lineplot(x='Metric', y='Value', hue='Cluster', data=rfm_melt, palette='viridis')
plt.title('Snake Plot of RFM Clusters')
plt.show()
```
